```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
#%matplotlib qt
```

```python
def plot_image(image, title=''):
    """
    Muestra una imagen utilizando matplotlib.
    :param img: Imagen a mostrar.
    :param title: Título de la imagen.
    """
    
    if image is None:
        print("No se pudo cargar la imagen.")
    else:
        plt.figure(figsize=(6, 6))
        if len(image.shape) == 3 and image.shape[2] == 3:
            image = cv.cvtColor(image, cv.COLOR_BGR2RGB)
            plt.imshow(image)
        else:
            plt.imshow(image, cmap='gray')
        plt.axis('off')
        plt.title(title)
        plt.show()
```

```python
def matrix_roi(image, fraction, m, n):
    """
    Crea una matriz de enfoque de la imagen, dividiendo en bloques de ROI.
    :param image: Imagen de entrada.
    :param fraction: Fracción del tamaño de la imagen para cada bloque ROI.
    :param M: Número de filas de bloques.
    :param N: Número de columnas de bloques.
    :return: Matriz de bloques ROI.
    """
       
    h, w = image.shape[:2]
    roi_h = int(h * fraction)
    roi_w = int(w * fraction)

    if roi_h <= 0 or roi_w <= 0:
        raise ValueError("El tamaño del ROI resultó cero. Ajustar 'fraction'.")

    total_h = roi_h * (2 * m - 1)
    total_w = roi_w * (2 * n - 1)

    if total_h > h or total_w > w:
        raise ValueError(f"La grilla ({total_h}x{total_w}) no entra en la imagen ({h}x{w}).")

    start_y0 = (h - total_h) // 2
    start_x0 = (w - total_w) // 2

    blocks = []
    for i in range(m):
        row = []
        for j in range(n):
            start_y = start_y0 + i * 2 * roi_h
            start_x = start_x0 + j * 2 * roi_w
            roi = image[start_y:start_y + roi_h, start_x:start_x + roi_w]
            row.append(roi)
        blocks.append(np.hstack(row))

    return np.vstack(blocks)
```

```python
def show_matrix_roi(image, fraction, m, n):
    """
    Muestra una grilla de ROIs centrados en la imagen original.
    :param image: Imagen de entrada.
    :param fraction: Fracción del tamaño de la imagen para cada bloque ROI. 
    :param m: Número de filas de bloques.
    :param n: Número de columnas de bloques.
    :return: None
    """

    h, w = image.shape[:2]
    roi_h = int(h * fraction)
    roi_w = int(w * fraction)

    if roi_h <= 0 or roi_w <= 0:
        raise ValueError("El tamaño del ROI resultó cero. Ajustar 'fraction'.")

    total_h = roi_h * (2 * m - 1)
    total_w = roi_w * (2 * n - 1)

    if total_h > h or total_w > w:
        raise ValueError(f"La grilla ({total_h}x{total_w}) no entra en la imagen ({h}x{w}).")

    # Copia para dibujar encima
    image_copy = image.copy()
    if len(image_copy.shape) == 2:
        image_copy = cv.cvtColor(image_copy, cv.COLOR_GRAY2BGR)

    start_y0 = (h - total_h) // 2
    start_x0 = (w - total_w) // 2

    for i in range(m):
        for j in range(n):
            start_y = start_y0 + i * 2 * roi_h
            start_x = start_x0 + j * 2 * roi_w
            end_y = start_y + roi_h
            end_x = start_x + roi_w
            cv.rectangle(image_copy, (start_x, start_y), (end_x, end_y), (0, 255, 0), 2)

    # Mostrar la imagen con matplotlib
    plt.imshow(cv.cvtColor(image_copy, cv.COLOR_BGR2RGB))
    plt.title("ROI")
    plt.axis('off')
    plt.show()
```

```python
def quality_image_measure(image, fraction, m, n):
    """
    Calcula la medida de calidad de imagen basada en la transformada de Fourier.
    :param I: Imagen en escala de grises.
    :param fraction: Fracción del tamaño de la imagen para el ROI.
    :param m: Número de filas de bloques.
    :param n: Número de columnas de bloques.
    :return: Lista de medidas de calidad de imagen para cada canal.
    """ 
    
    FM = []  # Lista para almacenar las medidas de calidad de imagen
    channels = range(image.shape[2]) if len(image.shape) == 3 else [None]

    for channel in channels:
        #I = select_roi(image[:, :, channel], fraction) if channel is not None else select_roi(image, fraction)
        I = matrix_roi(image[:, :, channel], fraction, m, n) if channel is not None else matrix_roi(image, fraction, m, n)
        # Aplico la transformada
        F = np.fft.fft2(I)
        # Llevo la baja frec al origen para acomodar la visualizacion
        Fc = np.fft.fftshift(F)
        # Obtengo el modulo
        AF = np.abs(Fc)
        # Calculo el maximo de la componente de frecuencia
        M = np.max(AF)
        # Calculo el numer total de pixeles en F tal que pixel > 0.001 * M
        AF_threshold = 0.001 * M
        Th = np.sum(AF > AF_threshold)
        # Calculo el Image Quality Measure
        FM.append(Th / (I.shape[0] * I.shape[1]))
           
    return FM
```

```python
# Uso el índce MIS2 (Brenner) para calcular la calidad de imagen
def quality_image_measure_brenner(image, fraction, m, n):
    """
    Calcula la medida de calidad de imagen basada en el índice de Brenner.
    :param image: Imagen de entrada.
    :param fraction: Fracción del tamaño de la imagen para el ROI.
    :param m: Número de filas de bloques.
    :param n: Número de columnas de bloques.
    :return: Medida de calidad de imagen basada en el índice de Brenner.
    """
    
    gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
    I = matrix_roi(gray, fraction, m, n)

    # Cálculo del índice de Brenner directamente sobre el ROI centrado
    shifted = I[:, 2:]  # desplazamiento de 2 píxeles
    original = I[:, :-2]
    brenner = np.sum((shifted - original) ** 2)
    
    return brenner
```

```python
def quality_video_measure(cap, quality_function, fraction, m, n):
    """
    Calcula la medida de calidad de cada cuadro de un video.
    :param cap: Objeto VideoCapture de OpenCV.
    :param fraction: Fracción del tamaño de la imagen para el ROI.
    :return: Lista de medidas de calidad de cada cuadro, calidad máxima y frame del máximo.
    """
    
    qualities = []
    max_quality = -float('inf')
    best_frame = None
    best_frame_idx = -1
    frame_idx = 0
    
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        # Calcular la medida de calidad
        quality = quality_function(frame, fraction, m, n)
        qualities.append(quality)
        
        current_quality = max(quality) if isinstance(quality, (list, np.ndarray)) else quality
        
        if current_quality > max_quality:
            max_quality = current_quality
            best_frame = frame.copy()
            best_frame_idx = frame_idx
        
        frame_idx += 1
    
    if best_frame is not None:
        show_matrix_roi(best_frame, fraction, m, n)
    
    cap.release()

    return qualities, max_quality, best_frame_idx
```

```python
def plot_video_quality(cap, quality_function, fraction=1.0, m=1, n=1):
    """
    Grafica la medida de calidad de cada cuadro del video.
    :param cap: Objeto VideoCapture de OpenCV.
    :param fraction: Fracción del tamaño de la imagen para el ROI.
    """
    
    list_quality, max_quality, best_frame_idx = quality_video_measure(cap, quality_function, fraction, m, n)
    arr_quality = np.array(list_quality)
    # Grafico la medida de calidad de cada cuadro del video
    plt.figure(figsize=(8, 4))
    if quality_function == quality_image_measure_brenner:
        plt.plot(arr_quality, marker='o', linestyle='', color='purple', label='Esc. de Grises')
        idx_max = np.argmax(arr_quality)
        plt.axvline(x=idx_max, color='black', linestyle='--', label='Máxima calidad')
        plt.title(f'Medida de Calidad de Cada Cuadro del Video - Fracción de la imagen (ROI) {fraction*100}% - Brenner')
        plt.legend()
    else:
        plt.plot(arr_quality[:, 0], marker='o', linestyle='', color='blue', label='Canal Azul')
        plt.plot(arr_quality[:, 1], marker='o', linestyle='', color='green', label='Canal Verde')
        plt.plot(arr_quality[:, 2], marker='o', linestyle='', color='red', label='Canal Rojo')
        idx_max = np.argmax([max(q) for q in list_quality])
        plt.axvline(x=idx_max, color='black', linestyle='--', label='Máxima calidad')
        plt.title(f'Medida de Calidad de Cada Cuadro del Video - Fracción de la imagen (ROI) {fraction*100}%')
        plt.legend()
    plt.xlabel('Número de Cuadro')
    plt.ylabel('Medida de Calidad')
    plt.grid()
    plt.show()
    
    return max_quality, best_frame_idx
```

```python
def unsharp_mask(image, kernel_size=(5, 5), sigma=1.0, amount=5.0):
    image_float = image.astype(np.float32)
    blurred = cv.GaussianBlur(image_float, kernel_size, sigma)
    mask = image_float - blurred
    sharpened = cv.addWeighted(image_float, 1.0, mask, amount, 0)
    return np.clip(sharpened, 0, 255).astype(np.uint8)


def improve_video_quality(cap):
    """
    Mejora la calidad de un video aplicando un filtro de máscara de enfoque.
    :param cap: Objeto VideoCapture de OpenCV.
    :return: video mejorado.
    """
    fourcc = cv.VideoWriter_fourcc(*'mp4v')
    out = cv.VideoWriter('improved_video.mp4', fourcc, 30.0, (int(cap.get(3)), int(cap.get(4))))
    
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        # Aplico la máscara de enfoque
        improved_frame = unsharp_mask(frame)
        out.write(improved_frame)
    
    cap.release()
    out.release()
    cv.destroyAllWindows()
```

```python
# Importo video
cap = cv.VideoCapture('focus_video.mov')
# Aplico la mejora de calidad
improve_video_quality(cap)
# Importo el video mejorado
cap = cv.VideoCapture('improved_video.mp4')

# Muestro la medida de calidad del video mejorado (evaluo la calidad de toda la imagen)
max_quality, best_frame_idx = plot_video_quality(cap, quality_image_measure)

print(f'Medida de calidad máxima: {max_quality:.4f}')
print(f'Cuadro con máxima calidad: {best_frame_idx}')
```
